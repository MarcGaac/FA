Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average.

```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def generate_random_list(size):
    return [random.randint(0, 1000) for _ in range(size)]

def measure_execution_time(size):
    random_list = generate_random_list(size)
    start_time = time.time()
    sorted(random_list)
    end_time = time.time()
    return end_time - start_time

sizes = np.logspace(2, 4, 10).astype(int)
execution_times = [measure_execution_time(size) for size in sizes]

slope, intercept, r_value, p_value, std_err = linregress(np.log(sizes), np.log(execution_times))

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")

plt.plot(sizes, execution_times, 'o-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Input Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Python\'s sorted Method')
plt.show()
```
