1. words.txt


```python
import requests
import random

def download_words():
    url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'
    response = requests.get(url)

    if response.status_code == 200:
        with open('words.txt', 'wb') as f:
            f.write(response.content)
            print('Dataset saved to file.')
    else:
        print('Failed to download dataset.')

def get_word_list():
    with open('words.txt', 'r') as f:
        word_list = f.read().split()
    return word_list

def total_words_count(word_list):
    return len(word_list)

def longest_word(word_list):
    return max(word_list, key=len)

def words_starting_with(letter, word_list):
    return [word for word in word_list if word.startswith(letter)]

def words_containing_substring(substring, word_list):
    return [word for word in word_list if substring in word]

def palindrome_words(word_list):
    return [word for word in word_list if word == word[::-1]]

def letter_frequency(word_list):
    frequencies = {}
    for word in word_list:
        for letter in word:
            frequencies[letter] = frequencies.get(letter, 0) + 1
    return frequencies

def word_guessing_game():
    word_list = get_word_list()
    selected_word = random.choice(word_list)
    guessed_letters = set()
    max_attempts = 6
    attempts = 0

    def display_word():
        masked_word = ''.join(letter if letter in guessed_letters else '_' for letter in selected_word)
        print(f"Word: {masked_word}")

    while True:
        display_word()
        guess = input("Enter a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        guessed_letters.add(guess)

        if guess not in selected_word:
            attempts += 1
            print(f"Incorrect! Attempts remaining: {max_attempts - attempts}")

        if set(guessed_letters) == set(selected_word):
            print("Congratulations! You guessed the word.")
            break

        if attempts == max_attempts:
            print(f"Sorry, you've run out of attempts. The word was '{selected_word}'.")
            break

if __name__ == "__main__":
    download_words()

    word_list = get_word_list()

    print(f"Total number of words: {total_words_count(word_list)}")
    print(f"Longest word: {longest_word(word_list)}")

    letter_input = input("Enter a letter to count words starting with it: ").lower()
    starting_with_letter = words_starting_with(letter_input, word_list)
    print(f"Number of words starting with '{letter_input}': {len(starting_with_letter)}")

    substring_input = input("Enter a substring to find matching words: ")
    containing_substring = words_containing_substring(substring_input, word_list)
    print(f"Words containing '{substring_input}': {containing_substring}")

    palindrome_words_list = palindrome_words(word_list)
    print(f"Palindromes: {palindrome_words_list}")

    frequencies = letter_frequency(word_list)
    print("Letter frequencies:")
    for letter, count in frequencies.items():
        print(f"{letter}: {count}")

    word_guessing_game()
```

2. Write a program on the Fibonacci sequence using both iterations and recursion, and including rendering time to compare the performance of both approaches.

```python
import time

# Iterative approach
def fibonacci_iterative(n):
    if n <= 1:
        return n

    fib_prev, fib_current = 0, 1

    for _ in range(2, n + 1):
        fib_prev, fib_current = fib_current, fib_prev + fib_current

    return fib_current

# Recursive approach
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Measure time taken by a function
def measure_time(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    elapsed_time = end_time - start_time
    return result, elapsed_time

if __name__ == "__main__":
    n = int(input("Enter the value of n for Fibonacci sequence: "))

    # Iterative approach
    iterative_result, iterative_time = measure_time(fibonacci_iterative, n)
    print(f"Iterative Result for n={n}: {iterative_result}")
    print(f"Iterative Time: {iterative_time:.6f} seconds\n")

    # Recursive approach
    recursive_result, recursive_time = measure_time(fibonacci_recursive, n)
    print(f"Recursive Result for n={n}: {recursive_result}")
    print(f"Recursive Time: {recursive_time:.6f} seconds")
```
